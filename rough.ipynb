```python
import numpy as np

def gf2_swap_rows(M, i, j):
    M[[i, j]] = M[[j, i]]


def gf2_swap_cols(M, i, j):
    M[:, [i, j]] = M[:, [j, i]]


def gf2_row_add(M, src, dst):
    M[dst] = (M[dst] + M[src]) % 2


def stabilizer_standard_form(G):
    G = G.copy() % 2
    m, total_cols = G.shape
    N = total_cols // 2
    col_perm = list(range(total_cols))

    r=0
    for col in range(N):
        pivot_row = None

        for row in range(r, m):
            if G[row, col] == 1:
                pivot_row = row
                break
        if pivot_row is None:
            continue



```
